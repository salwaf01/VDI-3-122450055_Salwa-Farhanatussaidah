```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'datacbl2022 (1).xlsx'
data = pd.read_excel(file_path)

# wilayah yang akan dibandingkann
wilayah_terpilih = ['ACEH SINGKIL', 'ACEH SELATAN', 'ACEH TENGGARA', 'ACEH TIMUR', 'ACEH TENGAH', 'ACEH BARAT', 'ACEH BESAR']

# Filter data berdasarkan wilayah yang dipilih
data_filtered = data[data['Wilayah'].isin(wilayah_terpilih)]

# Set up the plot size and style
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Membuat barplot Indeks Kedalaman Kemiskinan berdasarkan Wilayah
sns.barplot(x='Wilayah', y='Indeks Kedalaman Kemiskinan', data=data_filtered, palette='viridis')

# Menambahkan judul dan label
plt.title('Perbandingan Indeks Kedalaman Kemiskinan pada Wilayah di Aceh', fontsize=16)
plt.xlabel('Wilayah', fontsize=12)
plt.ylabel('Indeks Kedalaman Kemiskinan', fontsize=12)

# Menampilkan barplot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```



* TUJUAN =  Tujuan dari Visualisasi ini adalah untuk membandingkan indeks kedalaman kemiskinan di beberapa wilayah di Aceh, khususnya Aceh Singkil, Aceh Selatan, Aceh Tenggara, Aceh Timur, Aceh Tengah, Aceh Barat, dan Aceh Besar. Visualisasi ini memberikan gambaran mengenai tingkat kemiskinan di setiap wilayah untuk memahami kondisi sosial ekonomi.
